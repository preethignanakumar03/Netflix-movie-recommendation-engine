```python
#Most Popular Genre (Highest count of ratings)
popular_genres = df.groupby('genres')['Rating_x'].count().sort_values(ascending=False)
print("Most Popular Genres:\n", popular_genres.head())
```

```python
#Most Liked Genre (Highest average rating)
liked_genres = df.groupby('genres')['Rating_x'].mean().sort_values(ascending=False)
print("\nMost Liked Genres (Average Rating):\n", liked_genres.head())
```

```python
movie_stats = df.groupby(['genres', 'Name']).agg(
    avg_rating=('Rating_x', 'mean'),
    vote_count=('Rating_x', 'count')
).reset_index()
```

```python
qualified_movies = movie_stats[movie_stats['vote_count'] > 50]
```

```python
best_per_genre = qualified_movies.sort_values('avg_rating', ascending=False).drop_duplicates('genres')
```

```python
#Best Suited Movie per Genre
print("Best Suited Movie per Genre:")
print(best_per_genre[['genres', 'Name', 'avg_rating']])
```

```python
#Updated Genre Ratings (Best and Worst)
genre_comparison = df.groupby('genres')['Rating_x'].mean().sort_values(ascending=False)

print(f"Best Rated Genre: {genre_comparison.idxmax()} ({genre_comparison.max():.2f})")
print(f"Worst Rated Genre: {genre_comparison.idxmin()} ({genre_comparison.min():.2f})")
```
